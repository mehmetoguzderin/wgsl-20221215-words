```python
from spellchecker import SpellChecker
from autocorrect import Speller
import pkg_resources
from symspellpy import SymSpell, Verbosity
import requests
import re
import string
from bs4 import BeautifulSoup
```

```python
spell = SpellChecker()
```

```python
speller = Speller(lang='en')
```

```python
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)
dictionary_path = pkg_resources.resource_filename("symspellpy", "frequency_dictionary_en_82_765.txt")
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
```

```python
def check(word):
    return (spell.candidates(word),
            speller(word),
            {i.term for i in sym_spell.lookup(word, Verbosity.CLOSEST)})
```

```python
check("test")
```

```python
import string
from collections import Counter
Counter().most_common()[::-1]
string.ascii_lowercase
```

```python
# wgsl_text = str(open("wgsl.bs").read()).lower()
wgsl_text = BeautifulSoup(requests.get('https://gpuweb.github.io/gpuweb/wgsl/').text, 'html.parser').get_text().lower()
wgsl_words = [i for i in "".join([i if i in string.ascii_lowercase else " " for i in wgsl_text]).split() if len(i) > 2]
```

```python
wgsl_counts = open("wgsl_counts.txt", "w")
for i in Counter(wgsl_words).most_common()[::-1]:
    wgsl_counts.write(str(i) + "\n")
Counter(wgsl_words).most_common()[::-1]
```

```python
errlike = []
for word in wgsl_words:
    checks = check(word)
    if checks[0]:
        if word not in checks[0]:
            errlike.append(word)
    else:
        errlike.append(word)
    if checks[1]:
        if word not in checks[1]:
            errlike.append(word)
    else:
        errlike.append(word)
    if checks[2]:
        if word not in checks[2]:
            errlike.append(word)
    else:
        errlike.append(word)
Counter(errlike).most_common()
```
